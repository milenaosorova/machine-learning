```python
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import numpy as np
from skimage import draw
import torch

uploaded = files.upload()
```

```python
def plot2d(data):
    plt.imshow(data)
    plt.show()

def plot3d(data):
    fig = plt.figure(figsize=(10, 10))
    grid = ImageGrid(fig, 111, nrows_ncols=(1, 3), axes_pad=0.02)
    for ax, im in zip(grid, data):
        ax.imshow(im)
    plt.show()

# input_data = np.zeros((128, 128), dtype=np.int16)
# start = (20, 30)
# extent = (50, 60)
# rr, cc = draw.rectangle(start, extent=extent, shape=input_data.shape)
# # rr, cc = draw.circle_perimeter(14, 14, radius=10, shape=input_data.shape)
# input_data[rr, cc] = 1
# # input_data = images[2][35].pixel_array.astype(np.int16)
# input_data = np.expand_dims(input_data, axis=0)
# print(input_data.shape)
# plot2d(input_data[0])

with Image.open('sunflower.jpg') as img:
  img.load()
img = np.array(img.convert("L"))

plt.imshow(img)
plt.show()

input_data = np.expand_dims(img, axis=0)

print(input_data.shape)
```

```python
from torch import nn

conv1 = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=(13, 13), stride=1)

input_tensor = torch.from_numpy(input_data).to(torch.float32)
print(input_tensor.size())

output_tensor = conv1(input_tensor)
print(output_tensor.size())

np_out = output_tensor.detach().numpy()
print(np_out.shape)

plot3d(np_out)
```

```python
with Image.open('geom_objects.jpg') as img2:
  img2.load()
img2 = np.array(img2.convert("L"))

plt.imshow(img2)
plt.show()

input_data2 = np.expand_dims(img2, axis=0)

print(input_data.shape)
```

```python
conv2 = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=(13, 13), stride=1)

input_tensor2 = torch.from_numpy(input_data2).to(torch.float32)
print(input_tensor2.size())

output_tensor2 = conv2(input_tensor2)
print(output_tensor2.size())

np_out2 = output_tensor2.detach().numpy()
print(np_out2.shape)

plot3d(np_out2)
```

```python
relu = nn.ReLU()

output_tensor3 = relu(output_tensor2)
print(output_tensor3.size())

np_out3 = output_tensor3.detach().numpy()
print(np_out3.shape)

plot3d(np_out3)
```

```python
sigmoid = nn.Sigmoid()

output_tensor4 = sigmoid(output_tensor2)
print(output_tensor4.size())

np_out4 = output_tensor4.detach().numpy()
print(np_out4.shape)

plot3d(np_out4)
```

```python
tanhh = nn.Tanh()

output_tensor5 = tanhh(output_tensor2)
print(output_tensor5.size())

np_out5 = output_tensor5.detach().numpy()
print(np_out5.shape)

plot3d(np_out5)
```

```python
selu = nn.SELU()

output_tensor6 = selu(output_tensor2)
print(output_tensor6.size())

np_out6 = output_tensor6.detach().numpy()
print(np_out6.shape)

plot3d(np_out6)
```
